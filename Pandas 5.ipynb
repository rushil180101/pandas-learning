## Working with Dates and Time series data

```python
import pandas as pd

df = pd.read_csv("date_time_series_data.csv")
df
```

```python
df.info()
```

```python
# We need to convert Date from str format to datetime object/format
# using pd.to_datetime() function

# We need to specify the format of dates currently in str format

# Format can be looked up in documentation

df['Date'] = pd.to_datetime(df['Date'],format="%Y-%m-%d %I-%p")
```

```python
# day_name() function is used to get the day on a particular date
df.loc[0,'Date'].day_name()
```

```python
# Time interval between any two given dates
pd.to_datetime('2021-04-05 10-PM',format='%Y-%m-%d %I-%p') - pd.to_datetime('2017-04-17 01-AM',format='%Y-%m-%d %I-%p')
```

```python
# Filter data to get data between any two given dates
filt = (df['Date']>=pd.to_datetime('2018',format="%Y")) & (df['Date']<pd.to_datetime('2020',format="%Y"))
df.loc[filt]
```

```python
# We want all the data for Jan and Feb 2020
filt = (df['Date']>=pd.to_datetime('2020-01-01 12-AM',format="%Y-%m-%d %I-%p")) & (df['Date']<=pd.to_datetime('2020-02-29 11-PM',format="%Y-%m-%d %I-%p"))
jf20_df = df.loc[filt]
jf20_df
```

```python
jf20_df['Close'].mean()
```

```python
# Resampling
# Currently, the data is measured every hour (sampled on hourly basis)
# Suppose we want to resample it on daily basis
# e.g. Get the high value of each day, i.e., maximum of all the hourly
# values of that day.
# Use resample() function
# Refer documentation for date/day/hour/etc formats.

# ** To do this, the index should be in datetime format
df.set_index('Date',inplace=True)
df['High'].resample('D').max()
```

```python
# Sample every week
# Use 'W'
df['High'].resample('W').max()
```

```python
# Applying resampling to entire dataframe
# Use agg() function

df.resample('D').agg({
    'Open': 'mean',
    'Close': 'mean',
    'High': 'max',
    'Low': 'min',
    'Volume': 'sum'
})
```
