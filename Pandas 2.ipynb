## Sorting data

```python
import pandas as pd

data1 = {
    'first_name': ["Ramesh","Suresh","Kamlesh","Govind","Adam","Byomkesh","Jainish"],
    'last_name': ["Shah","Patel","Soni","Joshi","Shah","Shah","Patel"],
    'age': [40,28,33,35,27,28,37],
    'salary': [30000.00,25000.00,25000.00,35000.00,33000.00,25000.00,27000.00]
}

data1_df = pd.DataFrame(data1)
data1_df
```

```python
# Sort rows according to salary
# Use sort_values() method
data1_df.sort_values(by='salary')
```

```python
# Sort in descending order of salary
data1_df.sort_values(by='salary',ascending=False)
```

```python
# Multiple or Mixed sorting in case of duplicate values in one of the 
# sorting elements.
# e.g. If the salary is same for two or more employees, then
# sort them according to first_name
data1_df.sort_values(by=['salary','first_name'],ascending=[False,True])
```

```python
# Sorting data of a single column
data1_df['first_name'].sort_values()
```
