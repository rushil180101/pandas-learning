```python
import pandas as pd

data1 = {
    "emp id": ["e001","e043","e007","e025"],
    "first name": ["Naman","Kapil","Ramesh","Govind"],
    "last name": ["Joshi","Prajapati","Patel","Shah"],
    "email": ["nj@gmail.com","kp@gmail.com","rp@gmail.com","gs@gmail.com"],
    "age": [45,26,38,51]
}

data1_df = pd.DataFrame(data1)
data1_df
```
