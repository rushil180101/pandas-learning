```python
# How many people in each country know Python (using simple filter method)
import numpy as np

def contains_python(string):
    if (string==np.nan) or (str(string).find('Python')==-1):
        return False
    else:
        return True    

filt = (df['Country'] == 'India')
india_df = df[filt]
know_python = india_df['LanguageWorkedWith'].apply(contains_python)
ikp_df = india_df[know_python]
```

```python
# Get the number of people who know Python (using country_grp and apply() method)

country_grp['LanguageWorkedWith'].str.contains('Python')
```

```python
# 'SeriesGroupBy' object has no attribute 'str'
# Therefore, we can use apply() method and it will work for that column 
# for each country.
# e.g. 1st --> apply() works on 'LanguageWorkedWith' of 'Afghanistan'
#      2nd --> apply() works on 'LanguageWorkedWith' of 'Albania'
#              .........
#      nth --> apply() works on 'LanguageWorkedWith' of 'India'
#              .........

country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum()).loc[['United States','India']]
# Here, x takes on the Series 'LanguageWorkedWith' for each country group
```

```python
# What %age of people from each country know Python ?

# Method 1
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').value_counts(normalize=True)).head(30)
```

```python
# What %age of people from each country know Python ?

# Method 2

# Empty list
list1 = []

# Empyt dictionary
percentage_dict = {
    'Country': [],
    'Number of Respondents': [],
    'Number of Respondents who know Python': [],
    '%age of people who know Python': []
}

# Number of people who know Python in each country
country_know_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())

# Total number of people in each country
country_people = df['Country'].value_counts()

# Calculate percentage
for country in country_know_python.keys().tolist():
    people_know_python = country_know_python.loc[country]
    total_people = country_people.loc[country]
    percentage = (people_know_python/total_people)*100
    list1.append((country,total_people,people_know_python,percentage))
    
for element in list1:
    percentage_dict['Country'].append(element[0])
    percentage_dict['Number of Respondents'].append(element[1])
    percentage_dict['Number of Respondents who know Python'].append(element[2])
    percentage_dict['%age of people who know Python'].append(element[3])
    
percentage_df = pd.DataFrame(percentage_dict)
percentage_df.set_index('Country',inplace=True)
percentage_df
```
