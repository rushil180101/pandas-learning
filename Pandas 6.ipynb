## Reading/Writing Data to different sources like Excel,JSON,SQL,etc.

```python
import pandas as pd
```

```python
# Read csv file using read_csv()
df = pd.read_csv("survey_results_public.csv")
```

```python
df
```

```python
filt1 = (df['Country'] == "India")
filt2 = (df['Country'] == "United States")
india_df = df.loc[filt1]
us_df = df.loc[filt2]
```

```python
# Write to a file using to_csv() function
india_df.to_csv('india_survey_results.csv')
us_df.to_csv('united_states_survey_results.csv')
```

```python
# Read and Write to tab seperated files (tsv)
india_df.to_csv('india_survey_results_tsv.tsv',sep='\t')
india_tsv_df = pd.read_csv('india_survey_results_tsv.tsv',sep='\t')
india_tsv_df
```

```python
# To deal with excel files, we need 3 packages
# xlwt, openpyxl, xlrd
# Use to_excel() function
india_df.to_excel('india_survey_results_excel.xlsx')
```

```python
# Read from excel file using read_excel()
test_df = pd.read_excel('india_survey_results_excel.xlsx')
test_df
```

```python
test_df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
# Read/Write to JSON files
# to_json(), read_json()
india_df.to_json('india_survey_results_json.json')
test_df = pd.read_json('india_survey_results_json.json')
test_df
```
